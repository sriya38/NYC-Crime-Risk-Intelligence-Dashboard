```python
columns_to_keep = [
    'cmplnt_num',
    'cmplnt_fr_dt',
    'cmplnt_fr_tm',
    'ofns_desc',
    'law_cat_cd',
    'boro_nm',
    'prem_typ_desc',
    'latitude',
    'longitude',
    'susp_age_group',
    'susp_race',
    'susp_sex',
    'vic_age_group',
    'vic_race',
    'vic_sex'
]

df = df[columns_to_keep]
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.duplicated().sum()
```

```python
df['cmplnt_fr_dt'] = pd.to_datetime(
    df['cmplnt_fr_dt'],
    errors='coerce'
)
```

```python
df['cmplnt_fr_dt'].dtype
```

```python
df['year'] = df['cmplnt_fr_dt'].dt.year
df['month'] = df['cmplnt_fr_dt'].dt.month_name()
df['day_of_week'] = df['cmplnt_fr_dt'].dt.day_name()

df['hour'] = pd.to_datetime(
    df['cmplnt_fr_tm'],
    format='%H:%M:%S',
    errors='coerce'
).dt.hour
```

```python
df[['year','month','day_of_week','hour']].head()
```

```python
df.to_csv('nyc_crime_cleaned.csv', index=False)
```
